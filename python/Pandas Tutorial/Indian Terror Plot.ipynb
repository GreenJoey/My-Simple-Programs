```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
events = pd.read_csv("globalterrorismdb_0616dist.csv", encoding="ISO-8859-1", low_memory=False)

events = events[events["country_txt"] == "India"]

events["city"]
```

```python
country_cross = pd.crosstab(events.iyear, events.city)

country_cross.head(5)
```

```python
plt.figure(figsize=(30, 30))

for column in country_cross.columns:
    plt.plot(country_cross.index, country_cross[column])

plt.xlabel('Year')
plt.ylabel('Number of Attacks')
```
