```python
ebola = pd.read_csv('https://raw.githubusercontent.com/chendaniely/scipy-2017-tutorial-pandas/master/data/ebola_country_timeseries.csv')

ebola
```

```python
ebola_melt = pd.melt(ebola,
                     id_vars=['Date', 'Day'],
                     var_name='cd_country',
                     value_name='count'
                    )
```

```python
ebola_melt
```

```python
var_split = ebola_melt.cd_country.str.split('_')
```

```python
var_split.str.get(-1)
```

```python
ebola_melt['status'] = var_split.str.get(0)
ebola_melt['country'] = var_split.str.get(1)
```

```python
ebola_melt.head()
```

```python
var_split = ebola_melt.cd_country.str.split('_', expand=True)

var_split
```

```python
var_split.columns = ['status1', 'country1']

var_split.head()
```

```python
ebola_clean = pd.concat([ebola_melt, var_split], axis=1)

ebola_clean.head()
```

```python
weather = pd.read_csv('https://raw.githubusercontent.com/chendaniely/scipy-2017-tutorial-pandas/master/data/weather.csv')

weather
```

```python
weather_melt = weather.melt(
    id_vars=['id', 'year', 'month', 'element'],
    var_name='day',
    value_name='temp'
)
```

```python
weather_melt
```

```python
weather_tidy = weather_melt.pivot_table(
    index=['id', 'year', 'month', 'day'],
    columns='element',
    values='temp'
).reset_index()

weather_tidy
```
