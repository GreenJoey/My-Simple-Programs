<b><i> Merging Data </i></b>

```python
person = pd.read_csv('https://raw.githubusercontent.com/chendaniely/scipy-2017-tutorial-pandas/master/data/survey_person.csv')
site = pd.read_csv('https://raw.githubusercontent.com/chendaniely/scipy-2017-tutorial-pandas/master/data/survey_site.csv')
survey = pd.read_csv('https://raw.githubusercontent.com/chendaniely/scipy-2017-tutorial-pandas/master/data/survey_survey.csv')
visited = pd.read_csv('https://raw.githubusercontent.com/chendaniely/scipy-2017-tutorial-pandas/master/data/survey_visited.csv')
```

```python
visited_subset = visited.iloc[[0, 2, 6], :]
visited_subset
```

```python
# One to One merge
# The `left_on` parameter is used to mention the considered key in `left` datset
# The `right_on` parameter is used to mention the considered key in `right` datset
o2o = pd.merge(left=site, right=visited_subset,
              left_on='name', right_on='site')
o2o
```

```python
# Many to One merge
m2o = site.merge(right=visited, left_on='name', right_on='site')
m2o
```
