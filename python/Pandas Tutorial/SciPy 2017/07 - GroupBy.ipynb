```python
import numpy as np
```

```python
# Group the dataset by "continent" and get column "lifeExp"
# The aggregrate the values using `np.mean` function

gapminder.groupby('continent')['lifeExp'].agg(np.mean)
```

```python
gapminder.groupby('continent')['lifeExp'].agg(np.std)
```

```python
def my_mean(values):
    n = len(values)
    s = np.sum(values)
    return s / n
```

```python
gapminder.groupby('continent')['lifeExp'].agg(my_mean)
```

```python
gapminder.groupby('continent')['lifeExp'].agg([
    np.count_nonzero,
    np.mean,
    np.std
])
```
